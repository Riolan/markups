```python
from PIL import Image, ImageDraw
import math
import numpy as np
from noise import pnoise2 # Perlin noise generator
from IPython.display import display, Image as IPythonImage
import io

def generate_perlin_noise_map(width, height, scale, octaves, persistence, lacunarity, seed=None):
    """
    Generates a 2D Perlin noise map.

    Args:
        width (int): Width of the noise map.
        height (int): Height of the noise map.
        scale (float): Determines the "zoom" level of the noise. Smaller scale = larger features.
        octaves (int): Number of noise layers. More octaves add detail.
        persistence (float): How much each octave contributes to the overall shape. (amplitude)
        lacunarity (float): How much the frequency changes for each octave.
        seed (int, optional): Seed for reproducibility. Defaults to None (random).

    Returns:
        numpy.ndarray: A 2D array of noise values, normalized between 0 and 1.
    """
    if seed is not None:
        np.random.seed(seed)

    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            # Coordinates for noise generation. Scale them down to control frequency.
            # Adding a large offset can help avoid visible seams if you were tiling.
            nx = x / width * scale
            ny = y / height * scale
            
            # Generate Perlin noise value
            value = pnoise2(nx, ny, octaves=octaves, persistence=persistence, lacunarity=lacunarity, repeatx=1024, repeaty=1024, base=seed if seed is not None else 0)
            
            # Normalize noise to [0, 1] range (pnoise2 typically returns values around -0.5 to 0.5)
            noise_map[y, x] = (value + 0.5) # Shift to 0-1, then clamp or further normalize if needed
    
    # Further normalize to ensure it's precisely between 0 and 1
    min_noise = noise_map.min()
    max_noise = noise_map.max()
    if (max_noise - min_noise) > 0:
        noise_map = (noise_map - min_noise) / (max_noise - min_noise)
    else: # Handle case where noise is flat (e.g., all same value)
        noise_map = np.full_like(noise_map, 0.5) # Default to mid-height

    return noise_map

def map_noise_to_color(noise_value):
    """
    Maps a noise value (0-1) to an RGBA color for terrain.
    """
    if noise_value < 0.3: # Water/low ground
        # Blend from deep blue to light blue
        intensity = noise_value / 0.3
        r = int(0 * (1 - intensity) + 0 * intensity)
        g = int(0 * (1 - intensity) + 100 * intensity)
        b = int(100 * (1 - intensity) + 255 * intensity)
        return (r, g, b, 255)
    elif noise_value < 0.5: # Grassy hills
        # Blend from light blue to green
        intensity = (noise_value - 0.3) / 0.2
        r = int(0 * (1 - intensity) + 50 * intensity)
        g = int(100 * (1 - intensity) + 180 * intensity)
        b = int(255 * (1 - intensity) + 50 * intensity)
        return (r, g, b, 255)
    elif noise_value < 0.7: # Brown earth/rock
        # Blend from green to brown
        intensity = (noise_value - 0.5) / 0.2
        r = int(50 * (1 - intensity) + 139 * intensity)
        g = int(180 * (1 - intensity) + 69 * intensity)
        b = int(50 * (1 - intensity) + 19 * intensity)
        return (r, g, b, 255)
    else: # Mountain peaks (grey/white)
        # Blend from brown to grey/white
        intensity = (noise_value - 0.7) / 0.3
        r = int(139 * (1 - intensity) + 200 * intensity)
        g = int(69 * (1 - intensity) + 200 * intensity)
        b = int(19 * (1 - intensity) + 200 * intensity)
        return (r, g, b, 255)


def create_terrain_hex_array_for_notebook(
    image_width,
    image_height,
    base_hexagon_size,
    spacing,
    perlin_scale=0.1,  # Controls overall size of hills/valleys
    perlin_octaves=4,  # Number of detail layers
    perlin_persistence=0.5, # How much each octave affects the final value
    perlin_lacunarity=2.0,  # How much the frequency increases for each octave
    height_amplitude=100, # Max vertical displacement
    vertical_stretch_factor=0.5, # How much hexagons stretch vertically based on height difference
    outline_color=(0, 0, 0, 100), # Slightly transparent black outline
    line_width=1,
    seed=None,
    save_to_file=None # Optional: provide a filename (e.g., "terrain_hex_array.png")
):
    """
    Creates a terrain-like array of hexagons using Perlin noise for height and stretching.
    Displays in Jupyter Notebook.

    Args:
        image_width (int): The width of the output image.
        image_height (int): The height of the output image.
        base_hexagon_size (int): The base size of each hexagon (center to vertex).
        spacing (int): The spacing between hexagon centers.
        perlin_scale (float): Scale for the Perlin noise.
        perlin_octaves (int): Number of octaves for Perlin noise.
        perlin_persistence (float): Persistence for Perlin noise.
        perlin_lacunarity (float): Lacunarity for Perlin noise.
        height_amplitude (int): Maximum vertical displacement of hexagon centers.
        vertical_stretch_factor (float): Multiplier for how much hexagons stretch vertically.
        outline_color (tuple): RGBA tuple for the hexagon outline color.
        line_width (int): The width of the hexagon outline.
        seed (int, optional): Seed for Perlin noise reproducibility.
        save_to_file (str, optional): If provided, saves the image to this filename.
                                      Use '.png' for transparency. Defaults to None.

    Returns:
        PIL.Image.Image: The generated Pillow Image object.
    """

    img = Image.new("RGBA", (image_width, image_height), (0, 0, 0, 0))  # Transparent background
    draw = ImageDraw.Draw(img)

    # Calculate approximate number of hexagons needed for noise map resolution
    # Make the noise map slightly larger than the image to avoid edge issues
    noise_map_width = int(image_width / (base_hexagon_size * 1.5)) + 2
    noise_map_height = int(image_height / (base_hexagon_size * math.sqrt(3)/2 * 2)) + 2
    
    noise_map = generate_perlin_noise_map(
        noise_map_width, noise_map_height, perlin_scale,
        perlin_octaves, perlin_persistence, perlin_lacunarity, seed
    )

    # Calculate hexagon dimensions based on base size
    base_hex_width = base_hexagon_size * 2
    base_hex_height = math.sqrt(3) / 2 * base_hex_width

    # Calculate horizontal and vertical step for placement
    # Use slightly tighter packing to increase hexagon count
    h_step = base_hex_width * 3 / 4 + spacing
    v_step = base_hex_height + spacing

    row_offset = base_hex_width / 4 + spacing / 2 # Offset for alternating rows

    # Iterate to place hexagons
    row_idx = 0
    for row in np.arange(0, image_height + base_hex_height, v_step):
        col_offset = 0
        if row_idx % 2 != 0:  # Offset every other row
            col_offset = base_hex_width / 2 + spacing / 2

        col_idx = 0
        for col in np.arange(0, image_width + base_hex_width, h_step):
            x_base = col + col_offset + base_hexagon_size
            y_base = row + base_hexagon_size

            # Map cell coordinates to noise map coordinates
            noise_x = int(x_base / image_width * noise_map_width)
            noise_y = int(y_base / image_height * noise_map_height)
            
            # Clamp to noise map dimensions
            noise_x = min(noise_x, noise_map_width - 1)
            noise_y = min(noise_y, noise_map_height - 1)

            height_val = noise_map[noise_y, noise_x] # Value between 0 and 1

            # Adjust Y position based on height (higher noise = lower Y in image)
            # We want higher terrain to appear "higher up" visually, so subtract from base Y
            # Scaling height_amplitude by image_height provides relative displacement
            y_adjusted = y_base - (height_val * height_amplitude)

            # Adjust hexagon size/stretch based on height
            # Scale the vertical dimension (height of the hexagon)
            stretched_hex_height = base_hex_height * (1 + height_val * vertical_stretch_factor)
            stretched_hexagon_size_y = stretched_hex_height / math.sqrt(3) # Recalculate 'radius' for vertical stretch
            
            # The 'radius' for X remains the base_hexagon_size
            current_hexagon_size_x = base_hexagon_size
            
            # Define hexagon vertices for an *ellipse-like* hexagon to simulate stretching
            vertices = []
            for i in range(6):
                angle_deg = 60 * i - 30
                angle_rad = math.pi / 180 * angle_deg
                
                # Apply different scaling for x and y to stretch the hexagon
                x = x_base + current_hexagon_size_x * math.cos(angle_rad)
                y = y_adjusted + stretched_hexagon_size_y * math.sin(angle_rad)
                vertices.append((x, y))

            # Color based on height
            fill_color = map_noise_to_color(height_val)

            # Only draw if the hexagon is roughly within image bounds
            # This check is less strict now to allow for partial drawing at edges
            # For more precise clipping, you'd need to check each vertex or simplify the polygon
            if (min(v[0] for v in vertices) < image_width and max(v[0] for v in vertices) > 0 and
                min(v[1] for v in vertices) < image_height and max(v[1] for v in vertices) > 0):
                draw.polygon(vertices, fill=fill_color, outline=outline_color, width=line_width)

            col_idx += 1
        row_idx += 1

    if save_to_file:
        img.save(save_to_file)
        print(f"Terrain hexagon array saved as {save_to_file}")

    return img

# --- How to use this in a Jupyter Notebook cell ---

print("Generating and displaying Perlin noise terrain-like hexagon array...")

# Experiment with these parameters!
# Smaller base_hexagon_size and spacing will increase hexagon count and detail.
# Adjust perlin_scale, octaves, persistence, lacunarity for different terrain looks.
# height_amplitude controls how dramatic the height changes are.
# vertical_stretch_factor controls how much the hexagons deform.

terrain_img = create_terrain_hex_array_for_notebook(
    image_width=1000,
    image_height=700,
    base_hexagon_size=20, # Smaller hexagons
    spacing=2,          # Tighter spacing
    perlin_scale=0.08,    # Good for rolling hills
    perlin_octaves=6,   # More detail
    perlin_persistence=0.4, # Less impact from higher octaves
    perlin_lacunarity=2.5,  # Frequencies grow faster
    height_amplitude=150, # More dramatic height variations
    vertical_stretch_factor=0.8, # More vertical stretching
    outline_color=(0, 0, 0, 80), # Lighter, semi-transparent outline
    line_width=1,
    seed=42, # Consistent terrain
    save_to_file="perlin_hex_terrain.png" # Saves the image
)

display(terrain_img)

print("\nGenerating another terrain with different parameters (more mountainous)...")

mountain_terrain_img = create_terrain_hex_array_for_notebook(
    image_width=1200,
    image_height=800,
    base_hexagon_size=15, # Even smaller hexagons for more detail
    spacing=1,
    perlin_scale=0.04,   # Larger features (mountains)
    perlin_octaves=8,    # Lots of detail
    perlin_persistence=0.6, # Stronger higher octaves
    perlin_lacunarity=2.2,
    height_amplitude=250, # Very dramatic height
    vertical_stretch_factor=1.2, # Significant stretching
    outline_color=(0, 0, 0, 100),
    line_width=1,
    seed=123,
    save_to_file="perlin_hex_mountains.png"
)

display(mountain_terrain_img)
```

```python
from PIL import Image, ImageDraw
import math
import numpy as np
from noise import pnoise2 # Perlin noise generator
from IPython.display import display, Image as IPythonImage
import io

# --- Noise and Color Functions (Remain the same) ---
def generate_perlin_noise_map(width, height, scale, octaves, persistence, lacunarity, seed=None):
    """
    Generates a 2D Perlin noise map.
    """
    if seed is not None:
        np.random.seed(seed)

    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            nx = x / width * scale
            ny = y / height * scale
            
            value = pnoise2(nx, ny, octaves=octaves, persistence=persistence, lacunarity=lacunarity, repeatx=1024, repeaty=1024, base=seed if seed is not None else 0)
            
            noise_map[y, x] = (value + 0.5)
    
    min_noise = noise_map.min()
    max_noise = noise_map.max()
    if (max_noise - min_noise) > 0:
        noise_map = (noise_map - min_noise) / (max_noise - min_noise)
    else:
        noise_map = np.full_like(noise_map, 0.5)

    return noise_map


def map_noise_to_color(noise_value):
    """
    Maps a noise value (0-1) to an RGBA color for terrain.
    """
    if noise_value < 0.3: # Water/low ground
        intensity = noise_value / 0.3
        r = int(0 * (1 - intensity) + 0 * intensity)
        g = int(0 * (1 - intensity) + 100 * intensity)
        b = int(100 * (1 - intensity) + 255 * intensity)
        return (r, g, b, 255)
    elif noise_value < 0.5: # Grassy hills
        intensity = (noise_value - 0.3) / 0.2
        r = int(0 * (1 - intensity) + 50 * intensity)
        g = int(100 * (1 - intensity) + 180 * intensity)
        b = int(255 * (1 - intensity) + 50 * intensity)
        return (r, g, b, 255)
    elif noise_value < 0.7: # Brown earth/rock
        intensity = (noise_value - 0.5) / 0.2
        r = int(50 * (1 - intensity) + 139 * intensity)
        g = int(180 * (1 - intensity) + 69 * intensity)
        b = int(50 * (1 - intensity) + 19 * intensity)
        return (r, g, b, 255)
    else: # Mountain peaks (grey/white)
        intensity = (noise_value - 0.7) / 0.3
        r = int(139 * (1 - intensity) + 200 * intensity)
        g = int(69 * (1 - intensity) + 200 * intensity)
        b = int(19 * (1 - intensity) + 200 * intensity)
        return (r, g, b, 255)

# --- Core Hexagon Drawing Function using Axial Coordinates ---
def create_axial_terrain_hex_array_for_notebook(
    image_width,
    image_height,
    hexagon_side_length, # Side length R
    perlin_scale=0.1,
    perlin_octaves=4,
    perlin_persistence=0.5,
    perlin_lacunarity=2.0,
    height_amplitude=100,
    outline_color=(0, 0, 0, 80),
    line_width=1,
    seed=None,
    save_to_file=None
):
    """
    Creates a terrain-like array of perfectly edge-tiling hexagons using Perlin noise
    for height deformation applied to shared vertices, using axial coordinates.
    Displays in Jupyter Notebook.

    Args:
        image_width (int): The width of the output image.
        image_height (int): The height of the output image.
        hexagon_side_length (int): The length of one side of the hexagon (R).
        perlin_scale (float): Scale for the Perlin noise.
        perlin_octaves (int): Number of octaves for Perlin noise.
        perlin_persistence (float): Persistence for Perlin noise.
        perlin_lacunarity (float): Lacunarity for Perlin noise.
        height_amplitude (int): Maximum vertical displacement of vertices.
        outline_color (tuple): RGBA tuple for the hexagon outline color.
        line_width (int): The width of the hexagon outline.
        seed (int, optional): Seed for Perlin noise reproducibility.
        save_to_file (str, optional): If provided, saves the image to this filename.
                                      Use '.png' for transparency. Defaults to None.

    Returns:
        PIL.Image.Image: The generated Pillow Image object.
    """

    img = Image.new("RGBA", (image_width, image_height), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)

    R = hexagon_side_length # Radius (center to vertex) for flat-topped hexes
    
    # Pre-calculate constants for pixel conversion
    # For flat-topped hexes:
    # x = R * (3/2 * q)
    # y = R * (sqrt(3)/2 * q + sqrt(3) * r)
    # The hex_corners list has to be adjusted based on the specific orientation (flat-top)
    # Vertex offsets relative to the center (from Red Blob Games, flat-top)
    hex_vertex_offsets = [
        (R * math.cos(math.radians(angle)), R * math.sin(math.radians(angle)))
        for angle in [0, 60, 120, 180, 240, 300] # Adjust for flat-top starting at (R,0)
    ]

    # Noise map resolution (matches image dimensions for high detail)
    noise_map_width = image_width
    noise_map_height = image_height
    noise_map = generate_perlin_noise_map(
        noise_map_width, noise_map_height, perlin_scale,
        perlin_octaves, perlin_persistence, perlin_lacunarity, seed
    )

    # Cache for deformed vertex positions to ensure seamless tiling
    # Key: (round(ideal_x), round(ideal_y)) -> (deformed_x, deformed_y, noise_value)
    vertex_cache = {}

    def get_deformed_vertex(ideal_pixel_x, ideal_pixel_y):
        """
        Calculates the deformed position of a vertex based on Perlin noise.
        Caches results for efficiency.
        """
        cache_key = (round(ideal_pixel_x), round(ideal_pixel_y))
        if cache_key in vertex_cache:
            return vertex_cache[cache_key]

        # Map pixel coordinates to noise map coordinates
        noise_x = int((ideal_pixel_x / image_width) * noise_map_width)
        noise_y = int((ideal_pixel_y / image_height) * noise_map_height)
        
        # Clamp noise coordinates
        noise_x = max(0, min(noise_x, noise_map_width - 1))
        noise_y = max(0, min(noise_y, noise_map_height - 1))

        noise_val = noise_map[noise_y, noise_x]

        # Apply height deformation to the Y coordinate
        deformed_y = ideal_pixel_y - (noise_val * height_amplitude)
        
        result = (ideal_pixel_x, deformed_y, noise_val)
        vertex_cache[cache_key] = result
        return result

    # Determine the axial coordinate range to cover the image
    # Approximate bounds for q and r to ensure full coverage
    # Based on max pixel dimensions and hex size
    max_q = int(image_width / (R * 1.5)) + 2 # R * 1.5 is horizontal step of center
    max_r = int(image_height / (R * math.sqrt(3))) + 2 # R * sqrt(3) is vertical step of center

    # Loop through axial coordinates
    # Iterate through a diamond-shaped range of (q, r) that covers the rectangle
    # The exact ranges can be tricky to optimize, so we'll use a generous bounding box.
    q_start = -max_q // 2 - 2
    q_end = max_q // 2 + 2
    r_start = -max_r // 2 - 2
    r_end = max_r // 2 + 2

    # Offset to center the grid in the image
    center_offset_x = image_width / 2
    center_offset_y = image_height / 2

    for q in range(q_start, q_end):
        for r in range(r_start, r_end):
            # Calculate implied 's' coordinate to ensure q+r+s = 0
            s = -q - r

            # Check if this hexagon is within a reasonable cubical bounds (optional, for debugging)
            # if abs(q) > max_q or abs(r) > max_r or abs(s) > max(max_q, max_r):
            #     continue

            # Calculate the ideal pixel center of the current hexagon (q, r)
            # From Red Blob Games - flat-topped
            ideal_center_x = R * (3/2 * q) + center_offset_x
            ideal_center_y = R * (math.sqrt(3)/2 * q + math.sqrt(3) * r) + center_offset_y

            # If the center is far outside the image, skip to optimize
            if not (-R*2 < ideal_center_x < image_width + R*2 and \
                    -R*2 < ideal_center_y < image_height + R*2):
                continue
            
            # Get the six deformed vertices for this hexagon
            deformed_vertices_and_noise = []
            for dx, dy in hex_vertex_offsets:
                ideal_vertex_x = ideal_center_x + dx
                ideal_vertex_y = ideal_center_y + dy
                deformed_vertices_and_noise.append(
                    get_deformed_vertex(ideal_vertex_x, ideal_vertex_y)
                )
            
            # Extract just the (x, y) coordinates for drawing
            polygon_vertices = [(v[0], v[1]) for v in deformed_vertices_and_noise]
            
            # Calculate average noise for coloring the hexagon
            # Use the center's noise or average of vertices for color
            # Here, we'll use the noise from the center for coloring
            # To get center noise, we pass its ideal pixel coords to get_deformed_vertex
            _, _, center_noise_val = get_deformed_vertex(ideal_center_x, ideal_center_y)
            fill_color = map_noise_to_color(center_noise_val)

            draw.polygon(polygon_vertices, fill=fill_color, outline=outline_color, width=line_width)

    if save_to_file:
        img.save(save_to_file)
        print(f"Axial coordinate terrain hexagon array saved as {save_to_file}")

    return img




# --- How to use this in a Jupyter Notebook cell ---

print("Generating and displaying perfectly edge-tiling Perlin noise terrain-like hexagon array using axial coordinates...")

axial_terrain_img = create_axial_terrain_hex_array_for_notebook(
    image_width=1200,
    image_height=800,
    hexagon_side_length=25, # R: Distance from center to vertex, same as side length for flat-top
    perlin_scale=0.08,
    perlin_octaves=6,
    perlin_persistence=0.5,
    perlin_lacunarity=2.0,
    height_amplitude=150,
    outline_color=(0, 0, 0, 80),
    line_width=1,
    seed=1234, # New seed for Axial
    save_to_file="axial_perlin_hex_terrain.png"
)

display(axial_terrain_img)

print("\nGenerating another axial tiling terrain with different parameters (more rugged, smaller hexes)...")

axial_rugged_terrain_img = create_axial_terrain_hex_array_for_notebook(
    image_width=1500,
    image_height=900,
    hexagon_side_length=15, # Smaller hexes
    perlin_scale=0.04,
    perlin_octaves=8,
    perlin_persistence=0.6,
    perlin_lacunarity=2.5,
    height_amplitude=250,
    outline_color=(0, 0, 0, 100),
    line_width=1,
    seed=5678, # New seed for Axial
    save_to_file="axial_rugged_perlin_hex_terrain.png"
)

display(axial_rugged_terrain_img)
```

```python
from PIL import Image, ImageDraw
import math
import numpy as np
from noise import pnoise2 # Perlin noise generator
from IPython.display import display, Image as IPythonImage, clear_output
from IPython.display import display, Image as IPythonImage
import io
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual

# --- Noise and Color Functions (Remain the same) ---
def generate_perlin_noise_map(width, height, scale, octaves, persistence, lacunarity, seed=None):
    """
    Generates a 2D Perlin noise map.
    """
    if seed is not None:
        np.random.seed(seed)

    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            nx = x / width * scale
            ny = y / height * scale
            
            value = pnoise2(nx, ny, octaves=octaves, persistence=persistence, lacunarity=lacunarity, repeatx=1024, repeaty=1024, base=seed if seed is not None else 0)
            
            noise_map[y, x] = (value + 0.5)
     
    min_noise = noise_map.min()
    max_noise = noise_map.max()
    if (max_noise - min_noise) > 0:
        noise_map = (noise_map - min_noise) / (max_noise - min_noise)
    else:
        noise_map = np.full_like(noise_map, 0.5)

    return noise_map




def map_noise_to_color(noise_value):
    """
    Maps a noise value (0-1) to an RGBA color for terrain.
    """
    if noise_value < 0.3: # Water/low ground
        intensity = noise_value / 0.3
        r = int(0 * (1 - intensity) + 0 * intensity)
        g = int(0 * (1 - intensity) + 100 * intensity)
        b = int(100 * (1 - intensity) + 255 * intensity)
        return (r, g, b, 255)
    elif noise_value < 0.5: # Grassy hills
        intensity = (noise_value - 0.3) / 0.2
        r = int(0 * (1 - intensity) + 50 * intensity)
        g = int(100 * (1 - intensity) + 180 * intensity)
        b = int(255 * (1 - intensity) + 50 * intensity)
        return (r, g, b, 255)
    elif noise_value < 0.7: # Brown earth/rock
        intensity = (noise_value - 0.5) / 0.2
        r = int(50 * (1 - intensity) + 139 * intensity)
        g = int(180 * (1 - intensity) + 69 * intensity)
        b = int(19 * (1 - intensity) + 19 * intensity)
        return (r, g, b, 255)
    else: # Mountain peaks (grey/white)
        intensity = (noise_value - 0.7) / 0.3
        r = int(139 * (1 - intensity) + 200 * intensity)
        g = int(69 * (1 - intensity) + 200 * intensity)
        b = int(19 * (1 - intensity) + 200 * intensity)
        return (r, g, b, 255)

# --- Core Hexagon Drawing Function using Axial Coordinates ---
def create_axial_terrain_hex_array_for_notebook(
    image_width,
    image_height,
    hexagon_side_length, # Side length R
    perlin_scale,
    perlin_octaves,
    perlin_persistence,
    perlin_lacunarity,
    height_amplitude,
    outline_color=(0, 0, 0, 80),
    line_width=1,
    seed=None,
    save_to_file=None
):
    """
    Creates a terrain-like array of perfectly edge-tiling hexagons using Perlin noise
    for height deformation applied to shared vertices, using axial coordinates.
    Displays in Jupyter Notebook.

    Args:
        image_width (int): The width of the output image.
        image_height (int): The height of the output image.
        hexagon_side_length (int): The length of one side of the hexagon (R).
        perlin_scale (float): Scale for the Perlin noise.
        perlin_octaves (int): Number of octaves for Perlin noise.
        perlin_persistence (float): Persistence for Perlin noise.
        perlin_lacunarity (float): Lacunarity for Perlin noise.
        height_amplitude (int): Maximum vertical displacement of vertices.
        outline_color (tuple): RGBA tuple for the hexagon outline color.
        line_width (int): The width of the hexagon outline.
        seed (int, optional): Seed for Perlin noise reproducibility.
        save_to_file (str, optional): If provided, saves the image to this filename.
                                      Use '.png' for transparency. Defaults to None.

    Returns:
        PIL.Image.Image: The generated Pillow Image object.
    """

    img = Image.new("RGBA", (image_width, image_height), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)

    R = hexagon_side_length # Radius (center to vertex) for flat-topped hexes
     
    # Pre-calculate constants for pixel conversion
    # For flat-topped hexes:
    # x = R * (3/2 * q)
    # y = R * (sqrt(3)/2 * q + sqrt(3) * r)
    # The hex_corners list has to be adjusted based on the specific orientation (flat-top)
    # Vertex offsets relative to the center (from Red Blob Games, flat-top)
    hex_vertex_offsets = [
        (R * math.cos(math.radians(angle)), R * math.sin(math.radians(angle)))
        for angle in [0, 60, 120, 180, 240, 300] # Adjust for flat-top starting at (R,0)
    ]

    # Noise map resolution (matches image dimensions for high detail)
    noise_map_width = image_width
    noise_map_height = image_height
    noise_map = generate_perlin_noise_map(
        noise_map_width, noise_map_height, perlin_scale,
        perlin_octaves, perlin_persistence, perlin_lacunarity, seed
    )

    # Cache for deformed vertex positions to ensure seamless tiling
    # Key: (round(ideal_x), round(ideal_y)) -> (deformed_x, deformed_y, noise_value)
    vertex_cache = {}

    def get_deformed_vertex(ideal_pixel_x, ideal_pixel_y):
        """
        Calculates the deformed position of a vertex based on Perlin noise.
        Caches results for efficiency.
        """
        cache_key = (round(ideal_pixel_x), round(ideal_pixel_y))
        if cache_key in vertex_cache:
            return vertex_cache[cache_key]

        # Map pixel coordinates to noise map coordinates
        noise_x = int((ideal_pixel_x / image_width) * noise_map_width)
        noise_y = int((ideal_pixel_y / image_height) * noise_map_height)
         
        # Clamp noise coordinates
        noise_x = max(0, min(noise_x, noise_map_width - 1))
        noise_y = max(0, min(noise_y, noise_map_height - 1))

        noise_val = noise_map[noise_y, noise_x]

        # Apply height deformation to the Y coordinate
        deformed_y = ideal_pixel_y - (noise_val * height_amplitude)
         
        result = (ideal_pixel_x, deformed_y, noise_val)
        vertex_cache[cache_key] = result
        return result

    # Determine the axial coordinate range to cover the image
    # Approximate bounds for q and r to ensure full coverage
    # Based on max pixel dimensions and hex size
    max_q = int(image_width / (R * 1.5)) + 2 # R * 1.5 is horizontal step of center
    max_r = int(image_height / (R * math.sqrt(3))) + 2 # R * sqrt(3) is vertical step of center

    # Loop through axial coordinates
    # Iterate through a diamond-shaped range of (q, r) that covers the rectangle
    # The exact ranges can be tricky to optimize, so we'll use a generous bounding box.
    q_start = -max_q // 2 - 2
    q_end = max_q // 2 + 2
    r_start = -max_r // 2 - 2
    r_end = max_r // 2 + 2

    # Offset to center the grid in the image
    center_offset_x = image_width / 2
    center_offset_y = image_height / 2

    for q in range(q_start, q_end):
        for r in range(r_start, r_end):
            # Calculate implied 's' coordinate to ensure q+r+s = 0
            s = -q - r

            # Calculate the ideal pixel center of the current hexagon (q, r)
            # From Red Blob Games - flat-topped
            ideal_center_x = R * (3/2 * q) + center_offset_x
            ideal_center_y = R * (math.sqrt(3)/2 * q + math.sqrt(3) * r) + center_offset_y

            # If the center is far outside the image, skip to optimize
            if not (-R*2 < ideal_center_x < image_width + R*2 and \
                    -R*2 < ideal_center_y < image_height + R*2):
                continue
             
            # Get the six deformed vertices for this hexagon
            deformed_vertices_and_noise = []
            for dx, dy in hex_vertex_offsets:
                ideal_vertex_x = ideal_center_x + dx
                ideal_vertex_y = ideal_center_y + dy
                deformed_vertices_and_noise.append(
                    get_deformed_vertex(ideal_vertex_x, ideal_vertex_y)
                )
             
            # Extract just the (x, y) coordinates for drawing
            polygon_vertices = [(v[0], v[1]) for v in deformed_vertices_and_noise]
             
            # Calculate average noise for coloring the hexagon
            # Use the center's noise or average of vertices for color
            # Here, we'll use the noise from the center for coloring
            # To get center noise, we pass its ideal pixel coords to get_deformed_vertex
            _, _, center_noise_val = get_deformed_vertex(ideal_center_x, ideal_center_y)
            fill_color = map_noise_to_color(center_noise_val)

            draw.polygon(polygon_vertices, fill=fill_color, outline=outline_color, width=line_width)

    if save_to_file:
        img.save(save_to_file)
        print(f"Axial coordinate terrain hexagon array saved as {save_to_file}")

    return img



from ipywidgets import Output
output_area = Output()
def interactive_terrain_generator(
    side_length=25,
    perlin_scale=0.08,
    octaves=6,
    persistence=0.5,
    lacunarity=2.0,
    height_amplitude=150,
    seed=1234
):
    #with clear_output(wait=False): # Clear previous output before drawing new image
    with output_area: # Use the Output widget as context manager
        clear_output(wait=True) # This clear_output should now work within the Output context
        print("Generating interactive Perlin noise terrain...")
        img = create_axial_terrain_hex_array_for_notebook(
            image_width=1200,
            image_height=800,
            hexagon_side_length=side_length,
            perlin_scale=perlin_scale,
            perlin_octaves=octaves,
            perlin_persistence=persistence,
            perlin_lacunarity=lacunarity,
            height_amplitude=height_amplitude,
            outline_color=(0, 0, 0, 80),
            line_width=1,
            seed=seed
        )
        display(img)

print("Use the sliders below to adjust the terrain parameters:")
interact(
    interactive_terrain_generator,
    side_length=widgets.IntSlider(min=5, max=50, step=1, value=25, description='Hex Size (R)'),
    perlin_scale=widgets.FloatSlider(min=0.01, max=0.2, step=0.005, value=0.08, description='Perlin Scale'),
    octaves=widgets.IntSlider(min=1, max=10, step=1, value=6, description='Octaves'),
    persistence=widgets.FloatSlider(min=0.1, max=1.0, step=0.05, value=0.5, description='Persistence'),
    lacunarity=widgets.FloatSlider(min=1.5, max=3.5, step=0.1, value=2.0, description='Lacunarity'),
    height_amplitude=widgets.IntSlider(min=0, max=300, step=10, value=150, description='Height Amplitude'),
    seed=widgets.IntSlider(min=0, max=10000, step=1, value=1234, description='Seed')
)
display(output_area) 

axial_rugged_terrain_img = create_axial_terrain_hex_array_for_notebook(
    image_width=1500,
    image_height=900,
    hexagon_side_length=15, # Smaller hexes
    perlin_scale=0.04,
    perlin_octaves=8,
    perlin_persistence=0.6,
    perlin_lacunarity=2.5,
    height_amplitude=250,
    outline_color=(0, 0, 0, 100),
    line_width=1,
    seed=5678, # New seed for Axial
    save_to_file="axial_rugged_perlin_hex_terrain.png"
)

display(axial_rugged_terrain_img)
```

```python
from PIL import Image, ImageDraw
import math
import numpy as np
from noise import pnoise2 # Perlin noise generator
from IPython.display import display, Image as IPythonImage, clear_output
import io
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
from ipywidgets import Output, Image as WidgetImage # Import WidgetImage for smoother updates

# --- Noise and Color Functions (Remain the same) ---
def generate_perlin_noise_map(width, height, scale, octaves, persistence, lacunarity, seed=None):
    """
    Generates a 2D Perlin noise map.
    """
    if seed is not None:
        np.random.seed(seed)

    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            nx = x / width * scale
            ny = y / height * scale
            
            value = pnoise2(nx, ny, octaves=octaves, persistence=persistence, lacunarity=lacunarity, repeatx=1024, repeaty=1024, base=seed if seed is not None else 0)
            
            noise_map[y, x] = (value + 0.5)
    
    min_noise = noise_map.min()
    max_noise = noise_map.max()
    if (max_noise - min_noise) > 0:
        noise_map = (noise_map - min_noise) / (max_noise - min_noise)
    else:
        noise_map = np.full_like(noise_map, 0.5)

    return noise_map

def map_noise_to_color(noise_value):
    """
    Maps a noise value (0-1) to an RGBA color for terrain.
    """
    if noise_value < 0.3: # Water/low ground
        intensity = noise_value / 0.3
        r = int(0 * (1 - intensity) + 0 * intensity)
        g = int(0 * (1 - intensity) + 100 * intensity)
        b = int(100 * (1 - intensity) + 255 * intensity)
        return (r, g, b, 255)
    elif noise_value < 0.5: # Grassy hills
        intensity = (noise_value - 0.3) / 0.2
        r = int(0 * (1 - intensity) + 50 * intensity)
        g = int(100 * (1 - intensity) + 180 * intensity)
        b = int(255 * (1 - intensity) + 50 * intensity)
        return (r, g, b, 255)
    elif noise_value < 0.7: # Brown earth/rock
        intensity = (noise_value - 0.5) / 0.2
        r = int(50 * (1 - intensity) + 139 * intensity)
        g = int(180 * (1 - intensity) + 69 * intensity)
        b = int(50 * (1 - intensity) + 19 * intensity)
        return (r, g, b, 255)
    else: # Mountain peaks (grey/white)
        intensity = (noise_value - 0.7) / 0.3
        r = int(139 * (1 - intensity) + 200 * intensity)
        g = int(69 * (1 - intensity) + 200 * intensity)
        b = int(19 * (1 - intensity) + 200 * intensity)
        return (r, g, b, 255)

# --- Core Hexagon Drawing Function using Axial Coordinates (MODIFIED) ---
def create_axial_terrain_hex_array_for_notebook(
    image_width,
    image_height,
    hexagon_side_length, # Side length R
    perlin_scale,
    perlin_octaves,
    perlin_persistence,
    perlin_lacunarity,
    height_amplitude,
    plain_level,      # Parameter for plains threshold
    plain_flatness,   # Parameter for plains height reduction
    noise_map_resolution_factor=1, # NEW: Factor to reduce noise map resolution (1 means full resolution)
    outline_color=(0, 0, 0, 80),
    line_width=1,
    seed=None,
    save_to_file=None
):
    """
    Creates a terrain-like array of perfectly edge-tiling hexagons using Perlin noise
    for height deformation applied to shared vertices, using axial coordinates.
    Includes parameters for generating flatter plain areas.
    """

    img = Image.new("RGBA", (image_width, image_height), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)

    R = hexagon_side_length
    hex_vertex_offsets = [
        (R * math.cos(math.radians(angle)), R * math.sin(math.radians(angle)))
        for angle in [0, 60, 120, 180, 240, 300]
    ]

    # MODIFIED: Reduce noise map resolution to speed up Perlin noise generation
    noise_map_width = max(1, image_width // noise_map_resolution_factor)
    noise_map_height = max(1, image_height // noise_map_resolution_factor)

    noise_map = generate_perlin_noise_map(
        noise_map_width, noise_map_height, perlin_scale,
        perlin_octaves, perlin_persistence, perlin_lacunarity, seed
    )

    vertex_cache = {}

    def get_deformed_vertex(ideal_pixel_x, ideal_pixel_y):
        """
        Calculates the deformed position of a vertex based on Perlin noise.
        Caches results for efficiency. Incorporates plainness logic.
        """
        cache_key = (round(ideal_pixel_x), round(ideal_pixel_y))
        if cache_key in vertex_cache:
            return vertex_cache[cache_key]

        # MODIFIED: Map ideal pixel coordinates to the potentially smaller noise map coordinates
        noise_x = int((ideal_pixel_x / image_width) * noise_map_width)
        noise_y = int((ideal_pixel_y / image_height) * noise_map_height)
         
        noise_x = max(0, min(noise_x, noise_map_width - 1))
        noise_y = max(0, min(noise_y, noise_map_height - 1))

        noise_val_original = noise_map[noise_y, noise_x]

        current_height_amplitude = height_amplitude
        noise_val_for_color = noise_val_original # This will be the noise value used for color mapping

        # --- PLAINS LOGIC ---
        if noise_val_original < plain_level:
            t = noise_val_original / plain_level 
            reduction_factor = plain_flatness * (1 - t) + 1.0 * t
            current_height_amplitude = height_amplitude * reduction_factor
            
            # Optional: Remap noise_val for color in plains to make them more uniformly green.
            if noise_val_original < 0.5:
                noise_val_for_color = 0.3 + (noise_val_original / plain_level) * 0.2
                noise_val_for_color = max(0.0, min(1.0, noise_val_for_color))
        # --- END PLAINS LOGIC ---

        deformed_y = ideal_pixel_y - (noise_val_original * current_height_amplitude)
         
        result = (ideal_pixel_x, deformed_y, noise_val_for_color)
        vertex_cache[cache_key] = result
        return result

    # Determine the axial coordinate range to cover the image
    max_q = int(image_width / (R * 1.5)) + 2
    max_r = int(image_height / (R * math.sqrt(3))) + 2

    q_start = -max_q // 2 - 2
    q_end = max_q // 2 + 2
    r_start = -max_r // 2 - 2
    r_end = max_r // 2 + 2

    center_offset_x = image_width / 2
    center_offset_y = image_height / 2

    for q in range(q_start, q_end):
        for r in range(r_start, r_end):
            s = -q - r

            ideal_center_x = R * (3/2 * q) + center_offset_x
            ideal_center_y = R * (math.sqrt(3)/2 * q + math.sqrt(3) * r) + center_offset_y

            if not (-R*2 < ideal_center_x < image_width + R*2 and \
                    -R*2 < ideal_center_y < image_height + R*2):
                continue
            
            deformed_vertices_and_noise = []
            for dx, dy in hex_vertex_offsets:
                ideal_vertex_x = ideal_center_x + dx
                ideal_vertex_y = ideal_center_y + dy
                deformed_vertices_and_noise.append(
                    get_deformed_vertex(ideal_vertex_x, ideal_vertex_y)
                )
            
            polygon_vertices = [(v[0], v[1]) for v in deformed_vertices_and_noise]
            
            _, _, center_noise_val = get_deformed_vertex(ideal_center_x, ideal_center_y)
            fill_color = map_noise_to_color(center_noise_val)

            draw.polygon(polygon_vertices, fill=fill_color, outline=outline_color, width=line_width)

    if save_to_file:
        img.save(save_to_file)
        print(f"Axial coordinate terrain hexagon array saved as {save_to_file}")

    return img

## Interactive Controls (MODIFIED for WidgetImage and noise_map_resolution_factor)

# Create an Output widget to hold text output (e.g., "Generating..." message)
output_text_area = Output()
# Create an ipywidgets.Image widget to display the generated terrain image.
# This widget will be updated in place for smoother transitions.
terrain_widget_image = WidgetImage(width=1200, height=800) # Set initial size, can be adjusted via layout

def interactive_terrain_generator(
    side_length=25,
    perlin_scale=0.08,
    octaves=6,
    persistence=0.5,
    lacunarity=2.0,
    height_amplitude=150,
    plain_level=0.4,
    plain_flatness=0.2,
    noise_res_factor=2, # NEW: Slider parameter for noise map resolution factor
    seed=1234
):
    # Print messages to a dedicated output area
    with output_text_area:
        clear_output(wait=True) # Clear previous messages
        print("Generating interactive Perlin noise terrain...")

    # Generate the PIL Image
    pil_img = create_axial_terrain_hex_array_for_notebook(
        image_width=1200,
        image_height=800,
        hexagon_side_length=side_length,
        perlin_scale=perlin_scale,
        perlin_octaves=octaves,
        perlin_persistence=persistence,
        perlin_lacunarity=lacunarity,
        height_amplitude=height_amplitude,
        plain_level=plain_level,
        plain_flatness=plain_flatness,
        noise_map_resolution_factor=noise_res_factor, # Pass new parameter
        outline_color=(0, 0, 0, 80),
        line_width=1,
        seed=seed
    )

    # Convert PIL Image to bytes for ipywidgets.Image
    img_byte_arr = io.BytesIO()
    pil_img.save(img_byte_arr, format='PNG') # Using PNG as it supports transparency
    img_byte_arr = img_byte_arr.getvalue()

    # Update the value of the ipywidgets.Image widget.
    # This happens in-place and is much smoother than re-displaying.
    terrain_widget_image.value = img_byte_arr

print("Use the sliders below to adjust the terrain parameters:")

# Display the output widgets in the order you want them to appear.
# Display the image widget first, then the text output area.
display(terrain_widget_image)
display(output_text_area)

# Setup the interactive sliders.
interact(
    interactive_terrain_generator,
    side_length=widgets.IntSlider(min=5, max=50, step=1, value=25, description='Hex Size (R)'),
    perlin_scale=widgets.FloatSlider(min=0.01, max=0.2, step=0.005, value=0.08, description='Perlin Scale'),
    octaves=widgets.IntSlider(min=1, max=10, step=1, value=6, description='Octaves'),
    persistence=widgets.FloatSlider(min=0.1, max=1.0, step=0.05, value=0.5, description='Persistence'),
    lacunarity=widgets.FloatSlider(min=1.5, max=3.5, step=0.1, value=2.0, description='Lacunarity'),
    height_amplitude=widgets.IntSlider(min=0, max=300, step=10, value=150, description='Height Amplitude'),
    plain_level=widgets.FloatSlider(min=0.1, max=0.8, step=0.05, value=0.4, description='Plain Level (Noise)'),
    plain_flatness=widgets.FloatSlider(min=0.0, max=1.0, step=0.05, value=0.2, description='Plain Flatness (0=flat)'),
    noise_res_factor=widgets.IntSlider(min=1, max=10, step=1, value=2, description='Noise Res Factor'), # NEW Slider for performance
    seed=widgets.IntSlider(min=0, max=10000, step=1, value=1234, description='Seed')
);
```

```python
# --- IMPORTS ---
import math
import io
import numpy as np
from PIL import Image, ImageDraw
from noise import pnoise2  # Perlin noise generator

# Ipywidgets for interactive controls
import ipywidgets as widgets
from ipywidgets import interactive_output, VBox, HBox, Label
from ipywidgets import IntSlider, FloatSlider, Dropdown, ColorPicker,Button
from IPython.display import display, clear_output

# --- HELPER & CORE LOGIC FUNCTIONS ---

def _interpolate_color(start_rgb, end_rgb, factor):
    """Linearly interpolates between two RGB colors."""
    r = int(start_rgb[0] * (1 - factor) + end_rgb[0] * factor)
    g = int(start_rgb[1] * (1 - factor) + end_rgb[1] * factor)
    b = int(start_rgb[2] * (1 - factor) + end_rgb[2] * factor)
    return (r, g, b, 255)

def hex_to_rgb_tuple(hex_color):
    """Converts a hex color string (e.g., '#RRGGBB') to an RGB tuple (R, G, B)."""
    hex_color = hex_color.lstrip('#')
    if len(hex_color) != 6:
        return (0, 0, 0) # Return black for invalid hex
    return tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))

def generate_perlin_noise_map(width, height, scale, octaves, persistence, lacunarity, seed=None):
    """Generates a 2D Perlin noise map, normalized to the 0-1 range."""
    if seed is not None:
        np.random.seed(seed)
        base = seed
    else:
        base = 0

    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            nx = x / width * scale
            ny = y / height * scale
            noise_map[y, x] = pnoise2(nx, ny,
                                      octaves=octaves,
                                      persistence=persistence,
                                      lacunarity=lacunarity,
                                      repeatx=1024,
                                      repeaty=1024,
                                      base=base)

    # Normalize the noise map to be between 0 and 1
    min_noise = np.min(noise_map)
    max_noise = np.max(noise_map)
    return (noise_map - min_noise) / (max_noise - min_noise) if (max_noise - min_noise) > 0 else np.zeros_like(noise_map)


def map_noise_to_color(noise_value, color_mode="terrain",
                       custom_c0_3_hex='#000064', custom_c0_5_hex='#32B432',
                       custom_c0_7_hex='#8B4513', custom_c1_0_hex='#C8C8C8'):
    """Maps a noise value (0-1) to an RGBA color based on the selected mode."""
    if color_mode == "greyscale":
        intensity = int(noise_value * 255)
        return (intensity, intensity, intensity, 255)

    # Define color thresholds and corresponding colors for each mode
    if color_mode == "custom":
        thresholds = [0.3, 0.5, 0.7, 1.0]
        colors_rgb = [
            hex_to_rgb_tuple(custom_c0_3_hex),
            hex_to_rgb_tuple(custom_c0_5_hex),
            hex_to_rgb_tuple(custom_c0_7_hex),
            hex_to_rgb_tuple(custom_c1_0_hex),
            (255, 255, 255) # Final peak color (white)
        ]
    else: # Default to "terrain" mode
        thresholds = [0.3, 0.5, 0.7, 1.0]
        colors_rgb = [
            (0, 0, 100),    # Deep Water
            (0, 100, 255),  # Shallows
            (50, 180, 50),  # Grass/Land
            (139, 69, 19),  # Rock/Earth
            (200, 200, 200) # Mountain peaks
        ]

    # Find the correct color range and interpolate
    lower_bound = 0.0
    for i, upper_bound in enumerate(thresholds):
        if noise_value < upper_bound:
            # Normalize noise_value to a 0-1 factor within the current range
            factor = (noise_value - lower_bound) / (upper_bound - lower_bound) if (upper_bound - lower_bound) > 0 else 0
            start_color = colors_rgb[i]
            end_color = colors_rgb[i+1] if color_mode == 'terrain' else colors_rgb[i]
            if color_mode == 'custom': # Adjust logic for custom color interpolation
                 end_color = colors_rgb[i+1]

            # In terrain mode, we interpolate between adjacent colors.
            # In custom mode, the provided colors are for specific thresholds.
            if color_mode == "terrain":
                 return _interpolate_color(colors_rgb[i], colors_rgb[i+1], factor)
            else: # Custom and other modes
                if i == 0: return _interpolate_color(colors_rgb[0], colors_rgb[1], factor)
                if i == 1: return _interpolate_color(colors_rgb[1], colors_rgb[2], factor)
                if i == 2: return _interpolate_color(colors_rgb[2], colors_rgb[3], factor)
                if i == 3: return _interpolate_color(colors_rgb[3], colors_rgb[4], factor)


    return colors_rgb[-1] + (255,) # Return the highest color if all checks fail


def create_hex_terrain_image(
    image_width, image_height, hexagon_side_length,
    perlin_scale, perlin_octaves, perlin_persistence, perlin_lacunarity,
    height_amplitude, plain_level, plain_flatness,
    noise_map_resolution_factor=1,
    outline_color=(0, 0, 0, 80), line_width=1, seed=None,
    # Color parameters
    color_mode="terrain",
    custom_c0_3_hex='#000064', custom_c0_5_hex='#32B432',
    custom_c0_7_hex='#8B4513', custom_c1_0_hex='#C8C8C8'
):
    """
    Generates the final terrain image composed of deformed hexagons.
    This function contains the core drawing logic.
    """
    img = Image.new("RGBA", (image_width, image_height), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)

    R = hexagon_side_length
    hex_vertex_offsets = [
        (R * math.cos(math.radians(angle)), R * math.sin(math.radians(angle)))
        for angle in [0, 60, 120, 180, 240, 300]
    ]

    noise_map_width = max(1, image_width // noise_map_resolution_factor)
    noise_map_height = max(1, image_height // noise_map_resolution_factor)

    noise_map = generate_perlin_noise_map(
        noise_map_width, noise_map_height, perlin_scale,
        perlin_octaves, perlin_persistence, perlin_lacunarity, seed
    )

    vertex_cache = {}

    def get_deformed_vertex(ideal_pixel_x, ideal_pixel_y):
        cache_key = (round(ideal_pixel_x), round(ideal_pixel_y))
        if cache_key in vertex_cache:
            return vertex_cache[cache_key]

        noise_x = int((ideal_pixel_x / image_width) * noise_map_width)
        noise_y = int((ideal_pixel_y / image_height) * noise_map_height)
        noise_x = max(0, min(noise_x, noise_map_width - 1))
        noise_y = max(0, min(noise_y, noise_map_height - 1))
        
        noise_val = noise_map[noise_y, noise_x]

        # Apply plain flatness logic
        current_height_amplitude = height_amplitude
        if noise_val < plain_level:
            t = noise_val / plain_level if plain_level > 0 else 0
            reduction_factor = (1 - plain_flatness) * t + plain_flatness
            current_height_amplitude *= reduction_factor

        deformed_y = ideal_pixel_y - (noise_val * current_height_amplitude)
        
        result = (ideal_pixel_x, deformed_y, noise_val)
        vertex_cache[cache_key] = result
        return result

    # Calculate grid bounds
    max_q = int(image_width / (R * 1.5)) + 2
    max_r = int(image_height / (R * math.sqrt(3))) + 2
    q_range = range(-max_q // 2 - 2, max_q // 2 + 2)
    r_range = range(-max_r // 2 - 2, max_r // 2 + 2)
    center_offset_x = image_width / 2
    center_offset_y = image_height / 2

    for q in q_range:
        for r in r_range:
            ideal_center_x = R * (1.5 * q) + center_offset_x
            ideal_center_y = R * (math.sqrt(3) / 2 * q + math.sqrt(3) * r) + center_offset_y

            # Culling check to avoid drawing off-screen hexagons
            if not (-R*2 < ideal_center_x < image_width + R*2 and \
                    -R*2 < ideal_center_y < image_height + R*2):
                continue
            
            # Calculate deformed vertices and average noise for the polygon color
            polygon_vertices = []
            for dx, dy in hex_vertex_offsets:
                vx, vy, _ = get_deformed_vertex(ideal_center_x + dx, ideal_center_y + dy)
                polygon_vertices.append((vx, vy))

            _, _, center_noise_val = get_deformed_vertex(ideal_center_x, ideal_center_y)
            
            fill_color = map_noise_to_color(
                center_noise_val, color_mode,
                custom_c0_3_hex, custom_c0_5_hex, custom_c0_7_hex, custom_c1_0_hex
            )

            draw.polygon(polygon_vertices, fill=fill_color, outline=outline_color, width=line_width)
            
    return img

# --- INTERACTIVE WIDGETS SETUP ---

# 1. Define all widgets that will be used as controls
#    These are created only ONCE.

# -- Main Terrain Controls --
side_length_slider = IntSlider(min=5, max=50, step=1, value=25, description='Hex Size (R)')
perlin_scale_slider = FloatSlider(min=0.01, max=0.2, step=0.005, value=0.08, description='Perlin Scale', readout_format='.3f')
octaves_slider = IntSlider(min=1, max=10, step=1, value=6, description='Octaves')
persistence_slider = FloatSlider(min=0.1, max=1.0, step=0.05, value=0.5, description='Persistence')
lacunarity_slider = FloatSlider(min=1.5, max=3.5, step=0.1, value=2.0, description='Lacunarity')
height_amp_slider = IntSlider(min=0, max=300, step=10, value=150, description='Height Amp')
plain_level_slider = FloatSlider(min=0.1, max=0.8, step=0.05, value=0.4, description='Plain Level')
plain_flat_slider = FloatSlider(min=0.0, max=1.0, step=0.05, value=0.2, description='Plain Flatness')
noise_res_slider = IntSlider(min=1, max=10, step=1, value=2, description='Noise Res Factor')
seed_slider = IntSlider(min=0, max=10000, step=1, value=1234, description='Seed')

# -- Color Controls --
color_mode_dd = Dropdown(options=['terrain', 'greyscale', 'custom'], value='terrain', description='Color Mode:')
custom_c_water = ColorPicker(value='#000064', description='Water/Low')
custom_c_grass = ColorPicker(value='#32B432', description='Grass')
custom_c_rock = ColorPicker(value='#8B4513', description='Rock/Earth')
custom_c_mountain = ColorPicker(value='#C8C8C8', description='Mountain/Peak')

# This container for custom color pickers will be shown/hidden dynamically
custom_color_box = VBox([custom_c_water, custom_c_grass, custom_c_rock, custom_c_mountain])
custom_color_box.layout.display = 'none' # Hidden by default

# 2. Define the function that shows/hides the custom color pickers
def toggle_custom_colors(change):
    """Observes the color mode dropdown and toggles visibility of color pickers."""
    custom_color_box.layout.display = 'block' if change.new == 'custom' else 'none'

color_mode_dd.observe(toggle_custom_colors, names='value')

# 3. Define the main function that is called by the interactive controls
#    This function updates the output widgets (image and text area).
output_image = widgets.Image(width=1200, height=800)
output_log = widgets.Output()

def interactive_terrain_generator(
    side_length, perlin_scale, octaves, persistence, lacunarity,
    height_amplitude, plain_level, plain_flatness,
    noise_res_factor, seed, color_mode,
    custom_c0_3_hex, custom_c0_5_hex, custom_c0_7_hex, custom_c1_0_hex
):
    """
    This function is the target for interactive_output. It takes widget values,
    calls the image generation function, and updates the output widgets.
    """
    with output_log:
        clear_output(wait=True)
        print("Generating terrain...")

    pil_img = create_hex_terrain_image(
        image_width=1200, image_height=800,
        hexagon_side_length=side_length,
        perlin_scale=perlin_scale, perlin_octaves=octaves,
        perlin_persistence=persistence, perlin_lacunarity=lacunarity,
        height_amplitude=height_amplitude,
        plain_level=plain_level, plain_flatness=plain_flatness,
        noise_map_resolution_factor=noise_res_factor,
        seed=seed,
        color_mode=color_mode,
        custom_c0_3_hex=custom_c0_3_hex,
        custom_c0_5_hex=custom_c0_5_hex,
        custom_c0_7_hex=custom_c0_7_hex,
        custom_c1_0_hex=custom_c1_0_hex
    )

    # Convert the Pillow image to bytes and update the image widget
    with io.BytesIO() as img_byte_arr:
        pil_img.save(img_byte_arr, format='PNG')
        _latest_image_b64 = base64.b64encode(png_bytes).decode('utf-8')

        output_image.value = img_byte_arr.getvalue()

    with output_log:
        clear_output(wait=True)
        print("Done. Use the controls to adjust the terrain.")

# 4. Link the controls to the generation function using interactive_output
#    This creates a "headless" output that we will place in our final UI.
controls_dict = {
    'side_length': side_length_slider, 'perlin_scale': perlin_scale_slider,
    'octaves': octaves_slider, 'persistence': persistence_slider,
    'lacunarity': lacunarity_slider, 'height_amplitude': height_amp_slider,
    'plain_level': plain_level_slider, 'plain_flatness': plain_flat_slider,
    'noise_res_factor': noise_res_slider, 'seed': seed_slider,
    'color_mode': color_mode_dd,
    'custom_c0_3_hex': custom_c_water, 'custom_c0_5_hex': custom_c_grass,
    'custom_c0_7_hex': custom_c_rock, 'custom_c1_0_hex': custom_c_mountain,
}
interactive_out = interactive_output(interactive_terrain_generator, controls_dict)

# 5. Assemble the final UI layout
#    We arrange all our widget instances into a clean structure.
terrain_controls = VBox([
    side_length_slider, perlin_scale_slider, octaves_slider,
    persistence_slider, lacunarity_slider, height_amp_slider,
    plain_level_slider, plain_flat_slider, noise_res_slider, seed_slider
])

color_controls = VBox([
    color_mode_dd,
    custom_color_box # The container for the hide-able color pickers
])


def _copy(b):
    print(b)
    from IPython.display import display, clear_output, Javascript
    display(Javascript(f"""
        (async () => {{
            const button = document.querySelector('.p-Widget.jupyter-widgets.jupyter-button.widget-button[title="{b.tooltip}"]');
            try {{
                const blob = await fetch(`data:image/png;base64,{b}`).then(res => res.blob());
                await navigator.clipboard.write([ new ClipboardItem({{ 'image/png': blob }}) ]);
                button.innerText = 'Copied!';
            }} catch (err) {{
                console.error('Failed to copy image: ', err);
                button.innerText = 'Copy Failed!';
            }} finally {{
                setTimeout(() => {{ button.innerText = 'Copy Image'; }}, 2000);
            }}
        }})();
        """))

    


copy_button = Button(description="Copy Image", icon="clipboard", tooltip="Copies image to clipboard (may add background)")
copy_button.on_click(_copy)

# The final UI is a vertical box containing all other elements.
ui = VBox([
    Label("Use the sliders and controls below to adjust the terrain parameters:"),
    HBox([
        VBox([
            copy_button,
            Label("Terrain Controls:"),
            terrain_controls,
            Label("Color Controls:"),
            color_controls
        ]),
        VBox([
            output_image,
            output_log
        ])
    ]),
            # The 'headless' interactive_output is included here. It doesn't display
    # any controls itself, but it handles the communication between the
    # widgets and the function.
    interactive_out
])


# 6. Display the entire UI
#    This will show the controls and the output areas.
#    The interactive_output will automatically call the function on startup
#    to generate the initial image.
display(ui)
```

```python
# --- IMPORTS ---
import math
import io
import numpy as np
import base64  # <--- ADDED for encoding image data
from PIL import Image, ImageDraw
from noise import pnoise2
from ipywidgets import interact, interactive, fixed, interact_manual

# Ipywidgets for interactive controls
import ipywidgets as widgets
from ipywidgets import interactive_output, VBox, HBox, Label
from ipywidgets import IntSlider, FloatSlider, Dropdown, ColorPicker, Button # <--- ADDED Button
from IPython.display import display, clear_output, Javascript # <--- ADDED Javascript

# --- HELPER & CORE LOGIC FUNCTIONS (Unchanged) ---

def _interpolate_color(start_rgb, end_rgb, factor):
    """Linearly interpolates between two RGB colors."""
    r = int(start_rgb[0] * (1 - factor) + end_rgb[0] * factor)
    g = int(start_rgb[1] * (1 - factor) + end_rgb[1] * factor)
    b = int(start_rgb[2] * (1 - factor) + end_rgb[2] * factor)
    return (r, g, b, 255)

def hex_to_rgb_tuple(hex_color):
    """Converts a hex color string (e.g., '#RRGGBB') to an RGB tuple (R, G, B)."""
    hex_color = hex_color.lstrip('#')
    if len(hex_color) != 6:
        return (0, 0, 0) # Return black for invalid hex
    return tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))

def generate_perlin_noise_map(width, height, scale, octaves, persistence, lacunarity, seed=None):
    """Generates a 2D Perlin noise map, normalized to the 0-1 range."""
    if seed is not None:
        np.random.seed(seed)
        base = seed
    else:
        base = 0

    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            nx = x / width * scale
            ny = y / height * scale
            noise_map[y, x] = pnoise2(nx, ny,
                                      octaves=octaves,
                                      persistence=persistence,
                                      lacunarity=lacunarity,
                                      repeatx=1024,
                                      repeaty=1024,
                                      base=base)

    min_noise = np.min(noise_map)
    max_noise = np.max(noise_map)
    return (noise_map - min_noise) / (max_noise - min_noise) if (max_noise - min_noise) > 0 else np.zeros_like(noise_map)


def map_noise_to_color(noise_value, color_mode="terrain",
                       custom_c0_3_hex='#000064', custom_c0_5_hex='#32B432',
                       custom_c0_7_hex='#8B4513', custom_c1_0_hex='#C8C8C8'):
    """Maps a noise value (0-1) to an RGBA color based on the selected mode."""
    if color_mode == "greyscale":
        intensity = int(noise_value * 255)
        return (intensity, intensity, intensity, 255)

    if color_mode == "custom":
        thresholds = [0.3, 0.5, 0.7, 1.0]
        colors_rgb = [
            hex_to_rgb_tuple(custom_c0_3_hex), hex_to_rgb_tuple(custom_c0_5_hex),
            hex_to_rgb_tuple(custom_c0_7_hex), hex_to_rgb_tuple(custom_c1_0_hex),
            (255, 255, 255)
        ]
    else: # Default to "terrain" mode
        thresholds = [0.3, 0.5, 0.7, 1.0]
        colors_rgb = [
            (0, 0, 100), (0, 100, 255), (50, 180, 50),
            (139, 69, 19), (200, 200, 200)
        ]

    lower_bound = 0.0
    for i, upper_bound in enumerate(thresholds):
        if noise_value < upper_bound:
            factor = (noise_value - lower_bound) / (upper_bound - lower_bound) if (upper_bound - lower_bound) > 0 else 0
            if color_mode == "terrain":
                 return _interpolate_color(colors_rgb[i], colors_rgb[i+1], factor)
            else:
                if i == 0: return _interpolate_color(colors_rgb[0], colors_rgb[1], factor)
                if i == 1: return _interpolate_color(colors_rgb[1], colors_rgb[2], factor)
                if i == 2: return _interpolate_color(colors_rgb[2], colors_rgb[3], factor)
                if i == 3: return _interpolate_color(colors_rgb[3], colors_rgb[4], factor)

    return colors_rgb[-1] + (255,)


def create_hex_terrain_image(**kwargs):
    """Generates the final terrain image. Now accepts kwargs to simplify calls."""
    # Extract kwargs to keep the core logic clean
    image_width = kwargs.get('image_width')
    image_height = kwargs.get('image_height')
    # ... all other parameters ...
    
    img = Image.new("RGBA", (image_width, image_height), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)

    R = kwargs.get('hexagon_side_length')
    hex_vertex_offsets = [
        (R * math.cos(math.radians(angle)), R * math.sin(math.radians(angle)))
        for angle in [0, 60, 120, 180, 240, 300]
    ]

    noise_map = generate_perlin_noise_map(
        width=max(1, image_width // kwargs.get('noise_map_resolution_factor')),
        height=max(1, image_height // kwargs.get('noise_map_resolution_factor')),
        scale=kwargs.get('perlin_scale'), octaves=kwargs.get('perlin_octaves'),
        persistence=kwargs.get('perlin_persistence'), lacunarity=kwargs.get('perlin_lacunarity'),
        seed=kwargs.get('seed')
    )
    
    # ... The rest of the drawing logic is unchanged ...
    vertex_cache = {}
    def get_deformed_vertex(ideal_pixel_x, ideal_pixel_y):
        cache_key = (round(ideal_pixel_x), round(ideal_pixel_y))
        if cache_key in vertex_cache: return vertex_cache[cache_key]
        noise_x = int((ideal_pixel_x / image_width) * noise_map.shape[1])
        noise_y = int((ideal_pixel_y / image_height) * noise_map.shape[0])
        noise_x = max(0, min(noise_x, noise_map.shape[1] - 1))
        noise_y = max(0, min(noise_y, noise_map.shape[0] - 1))
        noise_val = noise_map[noise_y, noise_x]
        current_height_amplitude = kwargs.get('height_amplitude')
        if noise_val < kwargs.get('plain_level'):
            t = noise_val / kwargs.get('plain_level') if kwargs.get('plain_level') > 0 else 0
            reduction_factor = (1 - kwargs.get('plain_flatness')) * t + kwargs.get('plain_flatness')
            current_height_amplitude *= reduction_factor
        deformed_y = ideal_pixel_y - (noise_val * current_height_amplitude)
        result = (ideal_pixel_x, deformed_y, noise_val)
        vertex_cache[cache_key] = result
        return result
    max_q = int(image_width / (R * 1.5)) + 4; max_r = int(image_height / (R * math.sqrt(3))) + 4
    q_range = range(-max_q // 2, max_q // 2); r_range = range(-max_r // 2, max_r // 2)
    center_offset_x = image_width / 2; center_offset_y = image_height / 2
    for q in q_range:
        for r in r_range:
            ideal_center_x = R * (1.5 * q) + center_offset_x
            ideal_center_y = R * (math.sqrt(3) / 2 * q + math.sqrt(3) * r) + center_offset_y
            if not (-R*2 < ideal_center_x < image_width + R*2 and -R*2 < ideal_center_y < image_height + R*2): continue
            polygon_vertices = []
            for dx, dy in hex_vertex_offsets:
                vx, vy, _ = get_deformed_vertex(ideal_center_x + dx, ideal_center_y + dy)
                polygon_vertices.append((vx, vy))
            _, _, center_noise_val = get_deformed_vertex(ideal_center_x, ideal_center_y)
            fill_color = map_noise_to_color(center_noise_val, kwargs.get('color_mode'), kwargs.get('custom_c0_3_hex'), kwargs.get('custom_c0_5_hex'), kwargs.get('custom_c0_7_hex'), kwargs.get('custom_c1_0_hex'))
            draw.polygon(polygon_vertices, fill=fill_color, outline=kwargs.get('outline_color'), width=kwargs.get('line_width'))
    return img

# --- INTERACTIVE WIDGETS SETUP ---

# 1. Define all control widgets (unchanged)
side_length_slider = IntSlider(min=5, max=50, value=25, description='Hex Size (R)')
perlin_scale_slider = FloatSlider(min=0.01, max=0.2, step=0.005, value=0.08, description='Perlin Scale', readout_format='.3f')
# ... other sliders ...
octaves_slider = IntSlider(min=1, max=10, value=6, description='Octaves')
persistence_slider = FloatSlider(min=0.1, max=1.0, step=0.05, value=0.5, description='Persistence')
lacunarity_slider = FloatSlider(min=1.5, max=3.5, step=0.1, value=2.0, description='Lacunarity')
height_amp_slider = IntSlider(min=0, max=300, step=10, value=150, description='Height Amp')
plain_level_slider = FloatSlider(min=0.1, max=0.8, step=0.05, value=0.4, description='Plain Level')
plain_flat_slider = FloatSlider(min=0.0, max=1.0, step=0.05, value=0.2, description='Plain Flatness')
noise_res_slider = IntSlider(min=1, max=10, value=2, description='Noise Res Factor')
seed_slider = IntSlider(min=0, max=10000, value=1234, description='Seed')
color_mode_dd = Dropdown(options=['terrain', 'greyscale', 'custom'], value='terrain', description='Color Mode:')
custom_c_water = ColorPicker(value='#000064', description='Water/Low')
custom_c_grass = ColorPicker(value='#32B432', description='Grass')
custom_c_rock = ColorPicker(value='#8B4513', description='Rock/Earth')
custom_c_mountain = ColorPicker(value='#C8C8C8', description='Mountain/Peak')

custom_color_box = VBox([custom_c_water, custom_c_grass, custom_c_rock, custom_c_mountain])
custom_color_box.layout.display = 'none'

def toggle_custom_colors(change):
    custom_color_box.layout.display = 'block' if change.new == 'custom' else 'none'
color_mode_dd.observe(toggle_custom_colors, names='value')

# 2. Define output widgets and the NEW copy button
output_image = widgets.Image(width=1200, height=800)
output_log = widgets.Output()
copy_button = Button(description="Copy Image", icon="clipboard", tooltip="Copy the current image to the clipboard") # <--- NEW WIDGET
copy_button.layout.margin = '5px 0 0 0' # Add a little space above the button
copy_button.layout.width = '200px'

# 3. State variable and handler for the copy button
_latest_image_b64 = None # <--- NEW: Holds the latest image data for the copy function

def on_copy_button_click(b):
    """When the copy button is clicked, execute JavaScript to copy the image."""
    if _latest_image_b64:
        # This JS snippet uses the modern async clipboard API to write the image.
        # It's robust and asks for user permission if necessary.
        js_code = f"""
        (async () => {{
            const button = document.querySelector('.p-Widget.jupyter-widgets.jupyter-button.widget-button[title="{b.tooltip}"]');
            try {{
                const b64_data = '{_latest_image_b64}';
                const blob = await fetch(`data:image/png;base64,${{b64_data}}`).then(res => res.blob());
                await navigator.clipboard.write([
                    new ClipboardItem({{ 'image/png': blob }})
                ]);
                console.log('Image copied to clipboard!');
                button.innerText = 'Copied!';
                setTimeout(() => {{ button.innerText = 'Copy Image'; }}, 2000);
            }} catch (err) {{
                console.error('Failed to copy image: ', err);
                button.innerText = 'Copy Failed!';
                setTimeout(() => {{ button.innerText = 'Copy Image'; }}, 2000);
            }}
        }})();
        """
        display(Javascript(js_code))

copy_button.on_click(on_copy_button_click) # <--- Link handler to the button

# 4. The main generation function linked to interactive controls
def interactive_terrain_generator(**kwargs):
    """This function is the target for interactive_output."""
    global _latest_image_b64 # <--- Make sure we can write to the global variable
    
    with output_log:
        clear_output(wait=True)
        print("Generating terrain...")

    # Pass all kwargs to the generation function
    pil_img = create_hex_terrain_image(image_width=1200, image_height=800, **kwargs)

    # Convert the Pillow image to bytes for display and for copying
    with io.BytesIO() as img_byte_arr:
        pil_img.save(img_byte_arr, format='PNG')
        png_bytes = img_byte_arr.getvalue()
        
        # Update the image widget value
        output_image.value = png_bytes
        
        # --- NEW: Store the image data as Base64 for the copy function ---
        _latest_image_b64 = base64.b64encode(png_bytes).decode('utf-8')

    with output_log:
        clear_output(wait=True)
        print("Done. Use the controls to adjust the terrain.")

# 5. Link controls to the generation function (unchanged)
controls_dict = {
    'hexagon_side_length': side_length_slider, 'perlin_scale': perlin_scale_slider,
    'perlin_octaves': octaves_slider, 'perlin_persistence': persistence_slider,
    'perlin_lacunarity': lacunarity_slider, 'height_amplitude': height_amp_slider,
    'plain_level': plain_level_slider, 'plain_flatness': plain_flat_slider,
    'noise_map_resolution_factor': noise_res_slider, 'seed': seed_slider,
    'color_mode': color_mode_dd,
    'custom_c0_3_hex': custom_c_water, 'custom_c0_5_hex': custom_c_grass,
    'custom_c0_7_hex': custom_c_rock, 'custom_c1_0_hex': custom_c_mountain,
    # Pass some fixed parameters as well
    'outline_color': fixed((0, 0, 0, 80)), 'line_width': fixed(1)
}
interactive_out = interactive_output(interactive_terrain_generator, controls_dict)

# 6. Assemble the final UI layout, now including the copy button
terrain_controls = VBox([
    side_length_slider, perlin_scale_slider, octaves_slider,
    persistence_slider, lacunarity_slider, height_amp_slider,
    plain_level_slider, plain_flat_slider, noise_res_slider, seed_slider
])

color_controls = VBox([color_mode_dd, custom_color_box])

# The final UI now has the copy button placed under the image
ui = VBox([
    Label("Use the sliders and controls below to adjust the terrain parameters:"),
    HBox([
        VBox([
            Label("Terrain Controls:"), terrain_controls,
            Label("Color Controls:"), color_controls
        ]),
        VBox([ # This VBox contains the output area
            output_image,
            copy_button, # <--- ADDED BUTTON TO THE LAYOUT
            output_log
        ])
    ]),
    interactive_out
])

# 7. Display the UI
display(ui)
```

```python
def euclidean_dist(a1 : float, a2 : float, b1 : float, b2 : float) -> float:
    return ((a1 - b1)**2 + (a2 - b2)**2)**.5


def generate_perlin_noise_map(width, height, scale, octaves, persistence, lacunarity, seed=None):
    """Generates a 2D Perlin noise map, normalized to the 0-1 range."""
    if seed is not None:
        np.random.seed(seed)
        base = seed
    else:
        base = 0

    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            nx = x / width * scale
            ny = y / height * scale
            noise_map[y, x] = pnoise2(nx, ny,
                                      octaves=octaves,
                                      persistence=persistence,
                                      lacunarity=lacunarity,
                                      repeatx=1024,
                                      repeaty=1024,
                                      base=base)

    min_noise = np.min(noise_map)
    max_noise = np.max(noise_map)
    return (noise_map - min_noise) / (max_noise - min_noise) if (max_noise - min_noise) > 0 else np.zeros_like(noise_map)


import random
print(random.random())
pnts = []
JITTER = 0.5
for x in range(25):
    for y in range(25):
        pnts.append([x + JITTER * (random.random()
            - random.random()
            ), y + JITTER * (
             random.random()
            - random.random()
            )])


def drawer(canvas_size, pnts, size = 3):
    x,y = canvas_size
    image = Image.new("RGB", (x, y), "black")

    draw = ImageDraw.Draw(image)

    loc = 600/25
    pnts = [[x[0] * loc, x[1] * loc] for x in pnts]

    pnts = [[x[0] - size/2 , x[1] - size/2 , x[0] + size/2 , x[1] + size/2 ] for x in pnts]
    for pnt in pnts:
        draw.ellipse(pnt, fill="yellow")



    display(image)
drawer((600, 600), pnts)
```
